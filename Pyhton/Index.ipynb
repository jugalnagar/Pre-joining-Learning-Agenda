```python
#Create a function that grabs the email website domain from a string in the form: user@domain.com

def getDomain(str):
    index  = str.find('@')
    length = len(str);
    domain = str[index+1:length]
    return domain

print(getDomain('user@consultadd.com'))
```

```python
#  Create a function that counts the number of times the word "dog" occurs in a string. Again ignore edge cases.

str = 'This dog runs faster than the other dog dude!'

def countDog(str):
    list = str.split(" ");
    ans=0
    for strr in list:
        if strr=='dog':
            ans+=1;
    return ans

print(countDog(str))
```

```python
#Use lambda expressions and the filter() function to filter out words from a list that don't start 
#with the letter 's'. For example:seq = ['soup','dog','salad','cat','great']
#should be filtered down to:['soup','salad']
seq = ['soup','dog','salad','cat','great']
lst = filter(lambda str: str.startswith('s'),seq)

print(lst)

```

```python
#You are driving a little too fast, and a police officer stops you. Write a function
#to return one of 3 possible results: "No ticket", "Small ticket", or "Big Ticket". 
#If your speed is 60 or less, the result is "No Ticket". If speed is between 61 
#and 80 inclusive, the result is "Small Ticket". If speed is 81 or more, the result is "Big    Ticket". 
#Unless it is your birthday (encoded as a boolean value in the parameters of the function) -- on your birthday, 
#your speed can be 5 higher in all 

def caught_speeding(speed, is_birthday):
    if not is_birthday:
        if speed > 80:
            return "Big Ticket"
        elif speed > 60:
            return "Small Ticket"
        else :
            return "No Ticket"
    else:
        if speed > 85:
            return "Big Ticket"
        elif speed > 65:
            return "Small Ticket"
        else:
            return "No Ticket"
        

print(caught_speeding(81,True))

print(caught_speeding(81,False))

        
```
